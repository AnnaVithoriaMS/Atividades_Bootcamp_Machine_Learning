### Questão 01
Faça a leitura de uma imagem 2D colorida qualquer e apresente as
informações listadas abaixo por meio da função print().
altura, largura, Canais de cor, tipo de dado, tom de cinza máximo, tom de
cinza médio, tom de cinza mínimo

```python
plt.imshow(imagem)
plt.show
```

```python
if imagem is not None:
    altura = imagem.shape[0]
    largura = imagem.shape[1]
    canais = imagem.shape[2]
    tipo = imagem.dtype

    imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
    max = imagem_cinza.max()
    medio = imagem_cinza.mean()
    min = imagem_cinza.min()

    print(f'Altura : {altura}')
    print(f'Largura : {largura}')
    print(f'Canais de cor : {canais}')
    print(f'Tipo de dados : {tipo}')
    print(f'Tom máximo : {max}')
    print(f'Tom mínimo: {min}')
    print(f'Tom médio : {medio}')
```

### Questão 2

Faça a leitura de uma imagem de entrada 2D colorida qualquer e aplique o
filtro da média, mediana e gaussiano na imagem de entrada. Mostre os
resultados de cada filtro por meio da biblioteca matplotlib.

```python
#filtro da media
img_media = cv2.blur(imagem_cinza,(5,5))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(imagem_cinza,cv2.COLOR_BGR2RGB))
plt.title('Imagem Cinza')
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(img_media,cv2.COLOR_BGR2RGB))
plt.title('Imagem Suavizada - Filtro de Média')
```

```python
#filtro da mediana
img_mediana = cv2.medianBlur(imagem_cinza,5)
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(imagem_cinza,cv2.COLOR_BGR2RGB))
plt.title('Imagem Cinza')
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(img_mediana,cv2.COLOR_BGR2RGB))
plt.title('Imagem Suavizada - Filtro de Mediana')
```

```python
#filtro Gaussiano
img_gaussian = cv2.GaussianBlur(imagem_cinza,(7,7),0)
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(imagem_cinza,cv2.COLOR_BGR2RGB))
plt.title('Imagem Cinza')
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(img_gaussian,cv2.COLOR_BGR2RGB))
plt.title('Imagem Suavizada - Filtro Gaussiano')
```
